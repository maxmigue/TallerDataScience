```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta del archivo .log
ruta = "/content/drive/MyDrive/Utal/TALLER CIENCIA DE DATOS/Proyectos/Unidad 1/MHEALTH DATASET/mHealth_subject1.log"

# Definir nombres de columnas
columnas = [
    "chest_acc_x", "chest_acc_y", "chest_acc_z",
    "ecg_1", "ecg_2",
    "ankle_acc_x", "ankle_acc_y", "ankle_acc_z",
    "ankle_gyro_x", "ankle_gyro_y", "ankle_gyro_z",
    "ankle_mag_x", "ankle_mag_y", "ankle_mag_z",
    "arm_acc_x", "arm_acc_y", "arm_acc_z",
    "arm_gyro_x", "arm_gyro_y", "arm_gyro_z",
    "arm_mag_x", "arm_mag_y", "arm_mag_z",
    "label"
]

# Leer el archivo log (separado por espacios o tabs)
df = pd.read_csv(ruta, sep=r"\s+", header=None, names=columnas)

# Ver primeras filas
print(df.head())

# Ver cantidad de filas
print(df.shape)

# Ver tipo de variables
print(df.dtypes)

print("\n--- Medidas estad√≠sticas ---")
print(df.describe())

# 4. Distribuci√≥n de variables (histogramas)
df.hist(figsize=(15, 12), bins=30)
plt.suptitle("Distribuci√≥n de variables")
plt.show()

# 5. Correlaci√≥n entre variables
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False)
plt.title("Matriz de correlaci√≥n")
plt.show()

# 6. Visualizaci√≥n de algunas se√±ales (ejemplo)
plt.figure(figsize=(12, 6))
plt.plot(df["ecg_1"][:1000], label="ECG Lead 1")
plt.plot(df["ecg_2"][:1000], label="ECG Lead 2")
plt.title("Ejemplo de se√±al ECG (primeras 1000 muestras)")
plt.legend()
plt.show()
```
